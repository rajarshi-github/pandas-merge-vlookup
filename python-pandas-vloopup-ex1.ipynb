```python
import pandas as pd
```

```python
a = {
    'name'    : ["David", "James", "Prasanna", "Gulshan", "Debasis", "Joe"], 
    'pincode' : [500076,  500078,  400034, 100032, 700078, 10010]
    }

b = {
    'city_name': ["Bangalore","Mumbai", "Bangalore", "Chennai", "Delhi", "Kolkata", "Kolkata"], 
    'pincode'  : [500076, 200012 , 500078,  400034, 100032, 700078, 712304]
    }

df_a = pd.DataFrame(a)

df_b = pd.DataFrame(b)
```

```python
df_a
```

```python
df_b
```

```python
df_c = pd.merge(df_a, df_b, left_on='pincode', right_on='pincode' , how = 'inner')
```

```python
df_d = pd.merge(df_a, df_b, left_on='pincode', right_on='pincode', how='left')
```
